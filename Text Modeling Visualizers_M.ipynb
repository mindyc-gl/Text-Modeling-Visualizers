```python
df = pd.read_csv('covid_19_tweets.csv')
covid=df['tweet']
covid
```

```python
from nltk.tokenize import RegexpTokenizer
from stop_words import get_stop_words
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import pandas as pd

tokenizer = RegexpTokenizer(r'\w+')
# create English stop words list
en_stop = get_stop_words('en')
# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
```

```python
covid_list = list(covid)
texts = []
for i in covid_list:
     # clean and tokenize document string
     raw = i.lower()
     tokens = tokenizer.tokenize(raw)
     # remove stop words from tokens
     stopped_tokens = [i for i in tokens if not i in en_stop]
     # stem tokens
     stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
     # more than a letter
     short_tokens = [i for i in stemmed_tokens if len(i) > 2]

     # add tokens to list
     texts.append(short_tokens)

# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)

# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text, allow_update = True) for text in texts]

# Printing first 5 of DTM by word and frequency. To see full, use print(corpus)
for text in corpus[:5]:
     print([[dictionary[id], freq] for id, freq in text])
```

```python
booster_list = list(booster)
texts = []
for i in covid_list:
     # clean and tokenize document string
     raw = i.lower()
     tokens = tokenizer.tokenize(raw)
     # remove stop words from tokens
     stopped_tokens = [i for i in tokens if not i in en_stop]
     # stem tokens
     stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
     # more than a letter
     short_tokens = [i for i in stemmed_tokens if len(i) > 2]

     # add tokens to list
     texts.append(short_tokens)

# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)

# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text, allow_update = True) for text in texts]

# Printing first 5 of DTM by word and frequency. To see full, use print(corpus)
for text in corpus[:5]:
     print([[dictionary[id], freq] for id, freq in text])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# initialize a vectorizer
tfidfvect = TfidfVectorizer(analyzer='word', stop_words='english')

covid_list = list(covid)
booster_list = list(booster)

# converting docs into matrix
covid_tfidf = tfidfvect.fit_transform(covid_list)
booster_tfidf = tfidfvect.fit_transform(booster_list)
```

```python
#Apply PCA dimensionality reductions on the text
# For Covid-19 tweets
from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA

X = covid_tfidf
ipca = IncrementalPCA(n_components=2)
ipca.fit(X)
ipca.transform(X)
```

```python
#Apply PCA dimensionality reductions on the text
# For Booster shots tweets
from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA

X = booster_tfidf
ipca = IncrementalPCA(n_components=2)
ipca.fit(X)
ipca.transform(X)
```
